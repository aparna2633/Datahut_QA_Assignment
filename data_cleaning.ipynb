```python
import pandas as pd

df = pd.read_csv("messy_Data.csv") 

df.columns = df.columns.str.strip()

qa_issues = []

missing_values = df.isnull().sum()
for column in df.columns:
    if missing_values[column] > 0:
        missing_rows = df[df[column].isnull()].index.tolist()
        qa_issues.append([
            column, "Missing values found", "Missing Data", str(missing_rows),
            "Fill missing values with median/mode or remove", "Ensure data completeness during entry"
        ])

duplicates = df[df.duplicated()].index.tolist()
if duplicates:
    qa_issues.append([
        "Full Row", "Duplicate records found", "Duplicate Data", str(duplicates),
        "Remove duplicate rows", "Enforce unique record constraints"
    ])

if 'Email' in df.columns:
    invalid_emails = df[~df['Email'].str.match(r'^[\w\.-]+@[\w\.-]+\.\w{2,}$', na=False)].index.tolist()
    if invalid_emails:
        qa_issues.append([
            "Email", "Invalid email formats", "Field Value Error", str(invalid_emails),
            "Remove or correct invalid emails", "Use regex validation before storing"
        ])

if 'Email' in df.columns:
    non_professional_emails = df[~df['Email'].str.contains(r'@[a-zA-Z0-9]+\.(?:com|org|net)$', na=False, regex=True)].index.tolist()
    if non_professional_emails:
        qa_issues.append([
            "Email", "Non-professional email addresses", "Business Rule Violation", str(non_professional_emails),
            "Filter out non-corporate emails", "Ensure only business emails are collected"
        ])

if 'Join Date' in df.columns:
    df['Join Date'] = pd.to_datetime(df['Join Date'], errors='coerce')  
    invalid_dates = df[df['Join Date'].isnull()].index.tolist()
    if invalid_dates:
        qa_issues.append([
            "Join Date", "Incorrect date format", "Field Value Error", str(invalid_dates),
            "Convert to YYYY-MM-DD", "Use a uniform date format during data entry"
        ])

if 'Department' in df.columns:
    valid_departments = ["HR", "Engineering", "Marketing", "Sales", "Support"]
    invalid_departments = df[~df['Department'].isin(valid_departments)].index.tolist()
    if invalid_departments:
        qa_issues.append([
            "Department", "Typos in department names", "Field Value Error", str(invalid_departments),
            "Standardize department names", "Use predefined department list"
        ])

if 'Name' in df.columns:
    invalid_names = df[df['Name'].str.contains(r'[^a-zA-Z\s]', na=False, regex=True)].index.tolist()
    if invalid_names:
        qa_issues.append([
            "Name", "Names contain numbers/symbols", "Field Value Error", str(invalid_names),
            "Remove non-alphabetic characters", "Use regex validation on input"
        ])

if 'Salary' in df.columns:
    invalid_salaries = df[df['Salary'] <= 0].index.tolist()
    if invalid_salaries:
        qa_issues.append([
            "Salary", "Negative or zero salary values", "Field Value Error", str(invalid_salaries),
            "Remove or set minimum threshold", "Implement validation rules in data entry"
        ])

if 'Salary' in df.columns:
    salary_upper_bound = df['Salary'].quantile(0.95)  # 95th percentile as threshold
    outlier_salaries = df[df['Salary'] > salary_upper_bound].index.tolist()
    if outlier_salaries:
        qa_issues.append([
            "Salary", "Extremely high salary values", "Outlier Data", str(outlier_salaries),
            "Cap salary at a reasonable range", "Set upper/lower bounds for salary input"
        ])

if 'Employee ID' in df.columns:
    duplicate_employee_ids = df[df.duplicated(subset=['Employee ID'])].index.tolist()
    if duplicate_employee_ids:
        qa_issues.append([
            "Employee ID", "Duplicate Employee IDs", "Duplicate Data", str(duplicate_employee_ids),
            "Remove duplicate Employee IDs", "Enforce unique constraints in database"
        ])

qa_df = pd.DataFrame(qa_issues, columns=[
    "Field", "QA Issue", "Issue Type", "Data Line Number", "Solution", "Suggestion"
])

qa_df.to_excel("QA_Issues.xlsx", index=False)

print(" QA issues documented successfully in 'QA_Issues.xlsx' with exact row numbers.")
```

```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("messy_Data.csv")

df.fillna({
    'Salary': df['Salary'].median(),  
    'Department': 'Unknown',  
    'Email': 'missing@example.com',  
}, inplace=True)

df.dropna(subset=['Name', 'ID'], inplace=True)

df.drop_duplicates(inplace=True)

valid_email_pattern = r'^[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.(com|org|net)$'
df = df[df['Email'].str.match(valid_email_pattern, na=False)] 

df['Name'] = df['Name'].str.replace(r'[^a-zA-Z\s]', '', regex=True).str.title().str.strip()

df['Join Date'] = pd.to_datetime(df['Join Date'], errors='coerce', dayfirst=True).dt.strftime('%Y-%m-%d')

department_mapping = {
    'hr': 'HR', 'human resources': 'HR', 'engg': 'Engineering',
    'eng': 'Engineering', 'marketing': 'Marketing', 'sales': 'Sales',
    'support team': 'Support', 'customer support': 'Support'
}
df['Department'] = df['Department'].str.lower().map(department_mapping).fillna(df['Department'])

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')  
df = df[(df['Salary'] >= 15000) & (df['Salary'] <= 500000)] 

df.to_csv("cleaned_dataset.csv", index=False)
print("Data cleaning completed successfully!")
```
